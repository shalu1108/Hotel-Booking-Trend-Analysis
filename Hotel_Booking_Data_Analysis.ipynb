# **Data Analysis and Visualizations**

```python
# Count of Reservation Cancelled and Not-Cancelled
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5, 4))
plt.title('Reservation status count')

plt.bar(['Not canceled', 'Canceled'],
        df['is_canceled'].value_counts(),
        edgecolor='k',
        width=0.7,
        color=['#519d3e', '#ef8535'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette=['#519d3e', '#ef8535'])
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation status in different hotels', size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
# Seasonality – (Bookings Over Time)

df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'],
    categories=['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December'],
    ordered=True)

monthly_bookings = df.groupby(['arrival_date_year', 'arrival_date_month']).size().unstack('arrival_date_year').fillna(0)

plt.figure(figsize=(12, 6))
for year in monthly_bookings.columns:
    plt.plot(monthly_bookings.index, monthly_bookings[year], marker='o', label=year)
plt.title('Monthly Bookings by Year')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.show()
```

```python
# Revenue Insights – (Average Daily Rate by Hotel Type)
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='hotel', y='adr', ci=None, palette=sns.color_palette("Blues", n_colors=2))
plt.title('Average Daily Rate (ADR) by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('ADR (Euro)')
plt.show()
```

```python
# Cancelled and not cancelled percentage of each hotel..
resort_hotel = df.query("hotel == 'Resort Hotel'")
cancelled_perc = resort_hotel['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
```

```python
city_hotel = df.query("hotel == 'City Hotel'")
cancelled_perc = city_hotel['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
```

```python
# Customer Retention – (Repeated Guests)
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='is_repeated_guest', palette=['#A0CBE8', '#4E79A7'])
plt.title('Repeated vs Non-Repeated Guests')
plt.xlabel('Is Repeated Guest')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
# Grouping booking status date using groupby

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Visualization of average daily rate

plt.figure(figsize= (20,8))
plt.title('Average Daily Rate in City and Resort hotel', fontsize= 20)
plt.plot(resort_hotel.index, resort_hotel['adr'], label= 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label= 'City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
# Maximum and Minimum nunber of cancellations in months..
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette=['#6baed6', '#fdae6b'])
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation status per month', size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()

```

```python
# Average daily rate for each month..

colors = sns.color_palette("coolwarm", as_cmap=False)
adr_per_month = df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index()

plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=30)
sns.barplot(x='month', y='adr', data=adr_per_month,palette=colors)
plt.xlabel('Month')
plt.ylabel('ADR')
plt.show()
```

```python
# Top 10 countries by bookings
cancelled_data = df[df['is_canceled'] == 1]
top_countries = df['country'].value_counts().head(10)
colors = [
    '#4E79A7', '#A0CBE8', '#F28E2B', '#FFBE7D',
    '#D28B6C', '#9C755F', '#A9D18E',
    '#AF7AA1', '#FF9DA7', '#8E8883'
]
plt.figure(figsize=(8, 8))
plt.pie(
    top_countries,
    labels=top_countries.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1}
)
plt.title('Top 10 Countries by Bookings', fontsize=14, fontweight='semibold')
plt.tight_layout()
plt.show()
```

```python
# Cancellation Rate..
cancelled_data['market_segment'].value_counts(normalize = True)
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
```

```python
cancelled_df_adr = cancelled_df_adr[
    (cancelled_df_adr['reservation_status_date'] > '2016') &
    (cancelled_df_adr['reservation_status_date'] < '2017-09')
]
not_cancelled_df_adr =not_cancelled_df_adr[
    (not_cancelled_df_adr['reservation_status_date']>'2016') &
    (not_cancelled_df_adr['reservation_status_date']<'2017-09')
]
```

```python
# Average daily rate is the factor that is mostly influencing the cancellation rate

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate',fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 20)
plt.show()
```
